```python
device_keys = ['manufacturer_d_address_1', 'manufacturer_d_address_2', 'device_event_key', 'device_sequence_number',
                'manufacturer_d_state', 'manufacturer_d_zip_code', 'manufacturer_d_city', 'lot_number',
                'manufacturer_d_postal_code', 'manufacturer_d_zip_code_ext', 'date_received', 'model_number',
                'device_report_product_code', 'udi_public', 'device_operator', 'device_availability',
                'other_id_number', 'generic_name', 'manufacturer_d_name', 'brand_name', 'manufacturer_d_country',
                'combination_product_flag', 'openfda', 'device_age_text', 'udi_di',
                'device_evaluated_by_manufacturer', 'catalog_number', 'implant_flag', 'date_removed_flag']
mdr_text_keys = ['mdr_text_key', 'text_type_code', 'patient_sequence_number', 'text']
patient_keys = ['sequence_number_treatment', 'patient_sequence_number', 'patient_problems', 'date_received', 'sequence_number_outcome']
```

```python
def convert_list_of_dicts_to_dict(in_list, column_name):
    # print(in_list)
    try:
        out_dict = {column_name + "." + key: in_list[0][key] for key in in_list[0]}
        return out_dict
    except IndexError as error:
        if column_name.lower() == 'device':
            keys = device_keys
        elif column_name.lower() == 'mdr_text':
            keys = mdr_text_keys
        else:
            keys = patient_keys
        out_dict = {column_name + "." + key: 'NA' for key in keys}
        return out_dict
    except Exception as err:
        print(column_name)
        print(err)
        print(repr(err))
        raise Exception
```

```python
#  convert the entries in the device and mdr_text columns into Python dictionaries:
df['device'] = df['device'].apply(lambda dev: convert_list_of_dicts_to_dict(dev, "device"))
df['mdr_text'] = df['mdr_text'].apply(lambda mdr: convert_list_of_dicts_to_dict(mdr, "mdr_text"))
df['patient'] = df['patient'].apply(lambda mdr: convert_list_of_dicts_to_dict(mdr, "patient"))
```
