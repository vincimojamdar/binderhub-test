```python
device_keys = ['manufacturer_d_address_1', 'manufacturer_d_address_2', 'device_event_key', 'device_sequence_number',
               'manufacturer_d_state', 'manufacturer_d_zip_code', 'manufacturer_d_city', 'lot_number',
               'manufacturer_d_postal_code', 'manufacturer_d_zip_code_ext', 'date_received', 'model_number',
               'device_report_product_code', 'udi_public', 'device_operator', 'device_availability',
               'other_id_number', 'generic_name', 'manufacturer_d_name', 'brand_name', 'manufacturer_d_country',
               'combination_product_flag', 'openfda', 'device_age_text', 'udi_di',
               'device_evaluated_by_manufacturer', 'catalog_number', 'implant_flag', 'date_removed_flag']
mdr_text_keys = ['mdr_text_key', 'text_type_code', 'patient_sequence_number', 'text']
df['device'].fillna('{' + ''.join(key + '=, ' for key in device_keys) + '}', inplace=True)
df['mdr_text'].fillna('{' + ''.join(key + '=, ' for key in mdr_text_keys) + '}', inplace=True)
def convert_string_with_equal_to_sign_to_dict(in_string, column_name):
    out_dict = dict()
    in_string = str(in_string)
    in_string.strip()
    in_string = in_string[1:-1]
    if column_name == 'device':
        keys = device_keys
    else:
        keys = mdr_text_keys
    for idx, _ in enumerate(keys):
        key = column_name + "." + keys[idx]
        if idx < len(keys) - 1:
            search_pattern = keys[idx] + '=(.*?)' + keys[idx + 1] + '='
            result = re.search(search_pattern, in_string)
            if result:
                if result.group(1) == ', ':
                    out_dict[key] = None
                else:
                    out_dict[key] = result.group(1)
            else:
                out_dict[key] = None
    try:
        last_val = in_string.split(keys[-1] + '=')[1]
        if last_val == ', ':
            last_val = None
    except Exception as err:
        print(f"the error is: {repr(err)}")
        print(f"the in string is: {in_string}")
        raise Exception
    last_key = column_name + "." + keys[-1]
    out_dict[last_key] = last_val
    
    return out_dict
```
