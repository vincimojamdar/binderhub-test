```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X = np.random.randn(20000)
y = 10 + 9 * X + np.random.randn(20000)
```

```python
plt.scatter(X, y)
plt.show()
```

### Linear Regression using Analytical Approach (matrix mathematics)

```python
XX = np.c_[X, np.ones(20000)]
```

```python
beta = np.linalg.inv(XX.T @ XX) @ (XX.T @ y)
```

```python
beta
```

### Linear Regression using SKLearn's inbuild LinearRegression class; which shows that it does the same thing as the analytical solution:

```python
lin_reg = LinearRegression()
X_df = pd.DataFrame(X)
y_df = pd.DataFrame(y)
lin_reg.fit(X_df, y_df)
```

```python
lin_reg.coef_, lin_reg.intercept_
```

### Plotting the line over the data:

```python
y_pred = lin_reg.coef_ * X + lin_reg.intercept_
```

```python
plt.scatter(X, y)
plt.scatter(X, y_pred)
plt.show()
```
